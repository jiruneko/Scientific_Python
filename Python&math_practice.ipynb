# Python

```python
class Calc:
  def __init__(self, a):
    self.a = a

  def add(self, b):
    print(self.a + b)

  def multiply(self, b):
    print(self.a * b)
```

```python
cl = Calc(3)
cl.add(4)
cl.multiply(4)
```

```python
cl = Calc(4)

cl.add(5)
```

```python
class Calc:
  def __init__(self, a):
    self.a = a

  def __call__(self, c):
    print(self.a * c + c)
```

```python
cl = Calc(3)
cl(5)
```

```python
a = []
for i in range(0, 10):
  if i%2 == 1:
    a.append(i)

print(a)
```

```python
class Cat:
  def __init__(self, n, a):
    self.name = n
    self.age = a

cat1 = Cat('Tom', 3)
cat2 = Cat('Bob', 5)

print(cat1.name, cat1.age)
print(cat2.name, cat2.age)
```

```python
import numpy as np

x = np.linspace(-5, 5)

print(x)
print(len(x))
```

```python
import matplotlib.pyplot as plt

plt.plot(x, x**2)
```

```python
x = np.linspace(-5, 5)
y = 2*x**2 + 3*x + 1

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-3, 3)
y_1 = 1.5*x
y_2 = -2*x+1

print(y_1)
print(y_2)

# plt("x value", size=14)
plt.ylabel("y value", size=14)
plt.title("My Graph")

plt.grid()

plt.plot(x, y_1, label="y1")
plt.plot(x, y_2, label="y2")
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4)
y_1 = 2*x + 1
y_2 = x**2 - 4
y_3 = 0.5*x**3 - 6*x

plt.plot(x, y_1, label="1st")
plt.plot(x, y_2, label="2nd")
plt.plot(x, y_3, label="3rd")
```
